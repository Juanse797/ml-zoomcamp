```python
df = pd.read_csv('data.csv')
```

```python
#ajustar nombres de las columnas
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.sample(5)
```

```python
#crear una lista de las columnas de tipo object
string = list(df.dtypes[df.dtypes == 'object'].index)
string
```

```python
#cambiar los valores de las columnas de tipo object a minusculas y reemplazar espacios por guion bajo
for col in string:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
